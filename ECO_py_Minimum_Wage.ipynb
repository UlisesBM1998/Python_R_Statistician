```python
#Importing libraries
import numpy as np
import pandas as pd
```

```python
#Getting the data
dataset = pd.read_csv("njmin3.csv")
```

```python
#Describing the data set
pd.set_option("display.max_columns", None)
dataset.describe()
```

```python
#Checking if there are NA values
dataset.isnull().any()
```

```python
#Replacing NAs with averages
from sklearn.impute import SimpleImputer
missingValues = SimpleImputer(missing_values = np.nan, strategy = "mean")
missingValues = missingValues.fit(dataset[["fte", "demp"]])
dataset[["fte", "demp"]] = missingValues.transform(dataset[["fte", "demp"]])
dataset.isnull().any()
```

```python
#Isolating the X and Y variables
X = dataset.iloc[:,0:3].values
Y = dataset.iloc[:,3].values
```

```python
#Creating the first model
import statsmodels.api as sm
X = sm.add_constant(X)
model_a = sm.OLS(Y,X).fit()
model_a.summary(yname="FTE", xname = ("intercept", "New Jersey", "After April 1992",
                                     "New Jersey and after April 1992"))
```

```python
#Isolating the X and Y variables part 2
X = dataset.loc[:, ["NJ", "POST_APRIL92", "NJ_POST_APRIL92",
               "bk", "kfc", "wendys"]].values
```

```python
#Creating the second model
X = sm.add_constant(X)
model_b = sm.OLS(Y,X).fit()
model_b.summary(yname="FTE", xname = ("intercept", "New Jersey", "After April 1992",
                                     "New Jersey and after April 1992", "Burger King", "KFC", "Wendys"))
```

```python
#Isolating the X and Y variables part 3
X = dataset.loc[:, ["NJ", "POST_APRIL92", "NJ_POST_APRIL92",
               "bk", "kfc", "wendys", "co_owned", "centralj", "southj"]].values
```

```python
#Creating the third model
X = sm.add_constant(X)
model_c = sm.OLS(Y,X).fit()
model_c.summary(yname="FTE", xname = ("intercept", "New Jersey", "After April 1992",
                                     "New Jersey and after April 1992", "Burger King", "KFC", "Wendys", "Co-owned", "Central J",
                                     "South J"))
```
